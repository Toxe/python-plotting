# Histogram

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib_inline.backend_inline import set_matplotlib_formats

set_matplotlib_formats("svg")
```

```python
rg = np.random.default_rng()

values = rg.integers(0, 128, 8*8, dtype=np.uint8) + rg.integers(0, 128, 8*8, dtype=np.uint8)
values
```

```python
import matplotlib.patheffects as path_effects

text_outline = [path_effects.Stroke(linewidth=2, foreground="black"), path_effects.Normal()]

im = values.reshape((8, 8))
plt.imshow(im, cmap="gray", vmin=0, vmax=255, interpolation="none")

for y, row in enumerate(im):
    for x, color in enumerate(row):
        txt = plt.text(x, y, color, ha="center", va="center", color="white")
        txt.set_path_effects(text_outline)

plt.tight_layout()
plt.show()
```

## NumPy

```python
n, bins = np.histogram(values, bins=16)
list(zip(n, bins))
```

```python
n, bins = np.histogram(values, bins=16)
plt.plot(0.5 * (bins[1:] + bins[:-1]), n)

plt.tight_layout()
plt.show()
```

## Matplotlib

`plt.hist()` returns the histogram values and bins.

```python
n, bins, _ = plt.hist(values, bins=16)

plt.tight_layout()
plt.show()
```

```python
list(zip(n, bins))
```

### `range=`

Defines the range of values that the bins should cover. The default is the minimum and maximum value in `values`.

```python
plt.hist(values, bins=64, range=(0, 1000))

plt.tight_layout()
plt.show()
```

### `rwidth=`

Relative column width.

```python
plt.hist(values, rwidth=0.9)

plt.tight_layout()
plt.show()
```
