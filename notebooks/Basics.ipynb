# Basics

```python
from matplotlib import pyplot as plt
from matplotlib_inline.backend_inline import set_matplotlib_formats

set_matplotlib_formats("svg")
```

```python
from random import randrange

xvalues = list(range(20))
yvalues1 = [randrange(1, 101) for _ in range(len(xvalues))]
yvalues2 = [randrange(1, 101) for _ in range(len(xvalues))]
yvalues3 = [randrange(1, 101) for _ in range(len(xvalues))]
yvalues4 = [randrange(1, 101) for _ in range(len(xvalues))]
```

## Anatomy of a Figure

![anatomy of a figure](../images/anatomy_of_a_figure.webp)

## Single Axes

### Single Line (plot)

```python
plt.plot(xvalues, yvalues1, label="Values")

plt.xlabel("Index")
plt.ylabel("Random Values")
plt.title("Some Random Values")
plt.legend()

plt.tight_layout()
plt.show()
```

### Two Lines (plots) on one Axes

```python
plt.plot(xvalues, yvalues1, label="First")
plt.plot(xvalues, yvalues2, label="Second")

plt.xlabel("Index")
plt.ylabel("Random Values")
plt.title("Some Random Values")
plt.legend()

plt.tight_layout()
plt.show()
```
