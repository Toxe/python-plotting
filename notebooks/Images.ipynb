## Generate and Display Image Data

`vmin` and `vmax` specify the range of values within which the colors lie. For example, from 0 to 255 or 0.0 to 1.0.

`cmap="gray"` for grayscale images.

Differences between `imshow()` and `pcolormesh()`:

- With `pcolormesh()` the origin of the Y-axis is at the bottom left.
- `pcolormesh()` can display grid lines.
- `imshow()` always shows the pixels as square. For `pcolormesh()`, the scaling of the axes must be set with `plt.axis()` (either "image" or "scaled").

```python
rg = np.random.default_rng()

values = rg.integers(0, 256, (8, 8), dtype=np.uint8)
values
```

### .imshow()

`interpolation=` (e.g., "none" or "antialiased") specifies how the image data should be interpolated.

```python
plt.imshow(values, cmap="gray", vmin=0, vmax=255, interpolation="none")

plt.tight_layout()
plt.show()
```

#### `interpolation=`

```python
interpolations = [None, "none", "antialiased", "nearest", "bilinear", "bicubic", "spline16",
                  "spline36", "hanning", "hamming", "hermite", "kaiser", "quadric",
                  "catrom", "gaussian", "bessel", "mitchell", "sinc", "lanczos"]

plt.figure(figsize=(9, 12))

for i, interp in enumerate(interpolations):
    plt.subplot(5, 4, i+1, title=str(interp))
    plt.imshow(values, cmap="gray", vmin=0, vmax=255, interpolation=interp)
    plt.axis("off")

plt.tight_layout()
plt.show()
```

### .pcolormesh()

```python
plt.pcolormesh(values, cmap="gray", vmin=0, vmax=255, edgecolor="black")

plt.axis("image")
plt.tight_layout()
plt.show()
```
