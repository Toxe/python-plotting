# Axis

```python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib_inline.backend_inline import set_matplotlib_formats

set_matplotlib_formats("svg")
```

```python
x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)
```

## Hide Axis

```python
plt.figure(figsize=(3, 3))

plt.plot([-1, 1], [-1,  1], label="A")
plt.plot([-1, 1], [ 1, -1], label="B")

plt.axis("off")
plt.legend()

plt.tight_layout()
plt.show()
```

## Scaling

### Sine & cross from -1 to +1

```python
plt.figure(figsize=(8, 8))

for i, style in enumerate(["equal", "scaled", "tight", "auto", "image", "square"]):
    plt.subplot(3, 2, i+1, title=f'.axis("{style}")')
    plt.plot(x, y)
    plt.plot([-1, 1], [-1,  1])
    plt.plot([-1, 1], [ 1, -1])
    plt.axis(style)

plt.tight_layout()
plt.show()
```

### Sine & cross from -π to +π

```python
plt.figure(figsize=(8, 8))

for i, style in enumerate(["equal", "scaled", "tight", "auto", "image", "square"]):
    plt.subplot(3, 2, i+1, title=f'.axis("{style}")')
    plt.plot(x, y)
    plt.plot([-np.pi, np.pi], [-np.pi,  np.pi])
    plt.plot([-np.pi, np.pi], [ np.pi, -np.pi])
    plt.axis(style)

plt.tight_layout()
plt.show()
```

### 1:1 Image

```python
rg = np.random.default_rng()
data = rg.integers(0, 256, (8, 8), dtype=np.uint8)

plt.figure(figsize=(8, 8))

for i, style in enumerate(["equal", "scaled", "tight", "auto", "image", "square"]):
    plt.subplot(3, 2, i+1, title=f'.axis("{style}")')
    plt.pcolormesh(data, cmap="gray", vmin=0, vmax=255, edgecolor="black")
    plt.axis(style)

plt.tight_layout()
plt.show()
```
