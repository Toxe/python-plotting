# Subplots (Axes)

````python
import numpy as np
from matplotlib import pyplot as plt
from matplotlib_inline.backend_inline import set_matplotlib_formats

set_matplotlib_formats("svg")
````

````python
from random import randrange

xvalues = list(range(20))
yvalues1 = [randrange(1, 101) for _ in range(len(xvalues))]
yvalues2 = [randrange(1, 101) for _ in range(len(xvalues))]
yvalues3 = [randrange(1, 101) for _ in range(len(xvalues))]
yvalues4 = [randrange(1, 101) for _ in range(len(xvalues))]

x = np.linspace(-np.pi, np.pi, 100)
y = np.sin(x)
````

## Create Subplots

### plt.subplot()

With `nrows=2`, `ncols=2`, `index=1`:

- plt.subplot(2, 2, 1)
- plt.subplot(221)

````python
plt.figure(figsize=(3, 3))

for index in range(1, 5):
    plt.subplot(2, 2, index, title=f"22{index}")
    plt.plot(x, y)
    plt.axis("equal")

plt.tight_layout()
plt.show()
````

### plt.figure() + fig.add_subplot()

With `nrows=2`, `ncols=2`, `index=1`:

- fig.add_subplot(2, 2, 1)
- fig.add_subplot(221)

````python
fig = plt.figure(figsize=(3, 3))

for index in range(1, 5):
    ax = fig.add_subplot(2, 2, index, title=f"22{index}")
    ax.plot(x, y)
    ax.axis("equal")

plt.tight_layout()
plt.show()
````

### plt.subplots(nrows=, ncols=)

- plt.subplots(nrows=2, ncols=2)
- plt.subplots(2, 2)

Returns an array of Axes.

Use Tuple Unpacking for easier Axes object access.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)
fig, (ax_row1, ax_row2) = plt.subplots(2, 2)
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
```

````python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(3, 3))

axs[0][0].plot(x, y)
axs[0][1].plot(x, y)
axs[1][0].plot(x, y)
axs[1][1].plot(x, y)

axs[0][0].axis("equal")
axs[0][1].axis("equal")
axs[1][0].axis("equal")
axs[1][1].axis("equal")

plt.tight_layout()
plt.show()
````

### plt.subplots()

Returns a single Axes object.

````python
fig, ax = plt.subplots(figsize=(2, 2))

ax.plot(x, y)
ax.axis("equal")

plt.tight_layout()
plt.show()
````

## One row, two columns (1x2)

Example using `plt.subplot()`:

````python
plt.figure(figsize=(6, 2))

plt.subplot(121, title="First set of values")
plt.plot(xvalues, yvalues1)
plt.xlabel("Index")
plt.ylabel("Random Values")

plt.subplot(122, title="Second set of values")
plt.plot(xvalues, yvalues2)
plt.xlabel("Index")

plt.tight_layout()
plt.show()
````

## Two rows, one column (2x1)

Example using `fig.add_subplot()`:

````python
fig = plt.figure(figsize=(3, 4))

ax = fig.add_subplot(211, title="First set of values")
ax.plot(xvalues, yvalues1)
ax.set_ylabel("Random Values")

ax = fig.add_subplot(212, title="Second set of values")
ax.plot(xvalues, yvalues2)
ax.set_xlabel("Index")
ax.set_ylabel("Random Values")

plt.tight_layout()
plt.show()
````

## Two rows, two columns (2x2)

Example using `plt.subplots()`:

````python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(4, 4))

ax1.plot(xvalues, yvalues1)
ax2.plot(xvalues, yvalues2)
ax3.plot(xvalues, yvalues3)
ax4.plot(xvalues, yvalues4)

ax1.set_ylabel("Random Values")
ax1.set_title("Values #1")

ax2.set_title("Values #2")

ax3.set_xlabel("Index")
ax3.set_ylabel("Random Values")
ax3.set_title("Values #3")

ax4.set_xlabel("Index")
ax4.set_title("Values #4")

plt.tight_layout()
plt.show()
````

## Hide Axes

````python
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(4, 4))

axs[0][0].plot(xvalues, yvalues1)
axs[0][1].plot(xvalues, yvalues2)
axs[1][0].plot(xvalues, yvalues3)
axs[1][1].axis("off")  # hide

plt.tight_layout()
plt.show()
````
