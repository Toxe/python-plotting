# Styles

````python
from matplotlib import pyplot as plt
from matplotlib_inline.backend_inline import set_matplotlib_formats

set_matplotlib_formats("svg")
````

````python
from random import randrange

xvalues = list(range(20))
yvalues1 = [randrange(1, 101) for _ in range(len(xvalues))]
yvalues2 = [randrange(1, 101) for _ in range(len(xvalues))]
yvalues3 = [randrange(1, 101) for _ in range(len(xvalues))]
yvalues4 = [randrange(1, 101) for _ in range(len(xvalues))]
````

## .plot(..., ..., fmt)

https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html


fmt = `[marker][line][color]`, for example `"go-"`: green, circle marker, solid line

````python
plt.plot(xvalues, yvalues1, "go-", label="1st")   # green, circle marker, solid line
plt.plot(xvalues, yvalues2, "b.--", label="2nd")  # blue, point marker, dashed line
plt.plot(xvalues, yvalues3, "k:", label="3rd")    # black, dotted line
plt.plot(xvalues, yvalues4, "r*", label="4th")    # red, star marker

plt.legend()

plt.tight_layout()
plt.show()
````

## .plot() Line2D Properties

https://matplotlib.org/stable/api/_as_gen/matplotlib.lines.Line2D.html

### `alpha=`

````python
plt.plot(xvalues, yvalues1, label="1st", color="grey", alpha=0.1)
plt.plot(xvalues, yvalues2, label="2nd", color="grey", alpha=0.3)
plt.plot(xvalues, yvalues3, label="3rd", color="grey", alpha=0.7)
plt.plot(xvalues, yvalues4, label="4th", color="grey", alpha=1.0)

plt.legend()

plt.tight_layout()
plt.show()
````

### `color=` or `c=`

````python
plt.plot(xvalues, yvalues1, label="1st", color="green")
plt.plot(xvalues, yvalues2, label="2nd", color="blue")
plt.plot(xvalues, yvalues3, label="3rd", color="black")
plt.plot(xvalues, yvalues4, label="4th", color="#FF0000")

plt.legend()

plt.tight_layout()
plt.show()
````

### `drawstyle=` or `ds=`

````python
plt.plot(xvalues, yvalues1, "o--", color="grey", alpha=0.3)
plt.plot(xvalues, yvalues1, drawstyle="steps-pre")

plt.title("steps-pre (steps)")

plt.tight_layout()
plt.show()
````

````python
plt.plot(xvalues, yvalues1, "o--", color="grey", alpha=0.3)
plt.plot(xvalues, yvalues1, drawstyle="steps-mid")

plt.title("steps-mid")

plt.tight_layout()
plt.show()
````

````python
plt.plot(xvalues, yvalues1, "o--", color="grey", alpha=0.3)
plt.plot(xvalues, yvalues1, label="steps-post", drawstyle="steps-post")

plt.title("steps-post")

plt.tight_layout()
plt.show()
````

### `linestyle=` or `ls=`

````python
plt.plot(xvalues, yvalues1, label="1st", linestyle="solid")
plt.plot(xvalues, yvalues2, label="2nd", linestyle="dashed")
plt.plot(xvalues, yvalues3, label="3rd", linestyle="dotted")
plt.plot(xvalues, yvalues4, label="4th", linestyle="dashdot")

plt.legend()

plt.tight_layout()
plt.show()
````

### `linewidth=` or `lw=`

````python
plt.plot(xvalues, yvalues1, label="1st", linewidth=1)
plt.plot(xvalues, yvalues2, label="2nd", linewidth=2)
plt.plot(xvalues, yvalues3, label="3rd", linewidth=3)
plt.plot(xvalues, yvalues4, label="4th", linewidth=0, marker="*")

plt.legend()

plt.tight_layout()
plt.show()
````

### `marker=`

````python
plt.plot(xvalues, yvalues1, label="1st", marker=".")
plt.plot(xvalues, yvalues2, label="2nd", marker="o")
plt.plot(xvalues, yvalues3, label="3rd", marker="x")
plt.plot(xvalues, yvalues4, label="4th", marker="*")

plt.legend()

plt.tight_layout()
plt.show()
````

## plt.style

Either `plt.style.use("ggplot")` or as a Context Manager:

```python
with plt.style.context("ggplot", after_reset=True):
    # ...
    plt.show()
```

````python
plt.style.available
````

````python
for style in plt.style.available:
    with plt.style.context(style, after_reset=True):
        plt.plot(xvalues, yvalues1, label="1st")
        plt.plot(xvalues, yvalues2, label="2nd")
        plt.plot(xvalues, yvalues3, label="3rd")
        plt.plot(xvalues, yvalues4, label="4th")

        plt.xlabel("Index")
        plt.ylabel("Random Values")
        plt.title(style)
        plt.legend()

        plt.tight_layout()
        plt.show()
````
